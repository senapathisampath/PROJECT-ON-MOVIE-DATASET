```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("tmdb-movies.csv") ##loading the dataset
```

```python
df.head(4)##to print the first four index in the dataset
```

```python
#1-->Higher budget means Higher Popularity:
plt.scatter(df['budget'],df['popularity'])##scatter plot is used for the plotting
plt.title("Budget vs Popularity")## title for x and y axis
plt.xlabel("X-->Budget")## x axis is budget
plt.ylabel("Y-->Popularity")## y axis is popularity
```

From the above plot we can see that as Budget increses the popularity also increases.Hence Higher budget means Higher popularity.

```python
#2-->Is there Co-efficent relationship
df["budget"].corr(df["popularity"])
```

Hence they are positively correlated and there is coefficent relationship between budget and populartiy

```python
#3-->Will runtime affect votecount and popularity:
plt.scatter(df['runtime'],df['popularity'])## scatter plot between runtime and popularity
plt.title("runtime vs Popularity")
plt.xlabel("X-->runtime")
plt.ylabel("Y-->Popularity")

```

```python
plt.scatter(df['runtime'],df['vote_count'])## scatter plot between runtime and vote count
plt.title("runtime vs vote_count")
plt.xlabel("X-->runtime")
plt.ylabel("Y-->vote_count")
```

```python
df['profits']=df['revenue']-df['budget']##to calculate the profit
```

```python
#4-->Higher popularity means Higher profits:
plt.scatter(df['popularity'],df['profits'])##scatter plot between popularity and profit
plt.title("Popularity vs Profits")
plt.xlabel("X-->Popularity")
plt.ylabel("Y-->Profits")
```

From the above plot we can see that as popularity increses the profits also increases.Hence Higher popularity means Higher profits.

```python
#6-->Which genres are most popular from year to year

genre_details = list(map(str,(df['genres'])))
genre = ['Adventure', 'Science Fiction', 'Fantasy', 'Crime', 'Western', 'Family','nan', 'Animation','War','Mystery','Romance','TV Movie','Action', 'Thriller','Comedy','Drama' , 'History', 'Music', 'Horror', 'Documentary', 'Foreign']

year = np.array(df['release_year'])
popularity = np.array(df['popularity'])

popularity_df = pd.DataFrame(index = genre, columns = range(1960, 2016))
popularity_df = popularity_df.fillna(value = 0.0)

z = 0
for i in genre_details:
    split_genre = list(map(str,i.split('|')))
    popularity_df.loc[split_genre, year[z]] = popularity_df.loc[split_genre, year[z]] + popularity[z]
    z+=1
def calculate_std(x):
    return (x-x.mean())/x.std(ddof=0)

popular_genre = calculate_std(popularity_df)
popular_genre.head()
```
